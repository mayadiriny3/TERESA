EJERCICIO 1.- Desarrollar en serie de Fourier la función periodica de periodo 2π. Representar graficamente y estudiar la convergencia de la serie en 
$\mathbb{R}$  
$f(x) = \begin{cases} 
      0 & \text{si } -π \leq x \leq 0 \\
      x & \text{si } 0 \leq x \leq π  \\
   \end{cases}
$

```python
# Definir el rango de valores de x
import numpy as np
import matplotlib.pyplot as plt
x1 = np.linspace(-np.pi, 0, 1000)
x2 = np.linspace(0, np.pi, 1000)

# Definir la función f(x)
y1 = np.where(x1 <= 0, 0, x1)
y2 = x2

# Crear la figura y los ejes
fig, ax = plt.subplots()

# Graficar la función
ax.plot(x1, y1, color='blue', linewidth=2)
ax.plot(x2, y2, color='blue', linewidth=2)

# Etiquetas de los ejes
ax.set_xlabel('x')
ax.set_ylabel('f(x)')

# Título del gráfico
ax.set_title('Gráfica de f(x) = { 0 si -π ≤ x ≤ 0, x si 0 ≤ x ≤ π }')

# Mostrar la gráfica
plt.show()
```

#### Trasnformada de fourier rapida, discreta Graficar sonido, audio, representacion armonicos 10, 80, 160, 250, 500

```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

x = sp.Symbol('x')
n = sp.Symbol('n', integer=True)

# Definir la función periódica f(x)
f = sp.Piecewise((0, x < 0), (x, x >= 0))

# Calcular los coeficientes de Fourier para los primeros 10 armónicos
T = 2
armonicos = 10
a0 = sp.integrate(f, (x, -1, 1)) / T
an = [(2 / T) * sp.integrate(f * sp.cos(sp.pi * n * x), (x, -1, 1)) for n in range(1, armonicos + 1)]
bn = [(2 / T) * sp.integrate(f * sp.sin(sp.pi * n * x), (x, -1, 1)) for n in range(1, armonicos + 1)]


# Representación gráfica de la serie de Fourier
x_vals = np.linspace(-1, 1, 1000)
f_vals = np.piecewise(x_vals, [x_vals < 0, x_vals >= 0], [0, lambda x: x])
fourier_vals = a0/2 + sum([an_val * np.cos(n * np.pi * x_vals) + bn_val * np.sin(n * np.pi * x_vals)
                           for an_val, bn_val, n in zip(an, bn, range(1, armonicos + 1))])

plt.plot(x_vals, f_vals, label='Función original')
plt.plot(x_vals, fourier_vals, label='Serie de Fourier (10 armónicos)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

EJERCICIO 2.- Desarrollar en serie de Fourier la función periódica de período 2π, definida por:  
$f(x)=χ², -π \leq x \leq π $

<img src="https://drive.google.com/uc?id=1nnJHonHq-X60OgjorOjl9hauq0eS2IUL" alt="Menor a cero"  width="300" align="center"> <img src="https://drive.google.com/uc?id=16hTjMlf7hh_4MSFuln3Bb_zmaassY2p8" alt="Menor a cero"  width="300" align="center">

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función original
def f(x):
    return x**2

# Definir el número de términos de la serie truncada
num_terminos = 10

# Definir el intervalo de valores de x
x = np.linspace(-np.pi, np.pi, 1000)

# Calcular la serie truncada
serie_truncada = np.zeros_like(x)
for n in range(1, num_terminos+1):
    coeficiente_a = (4*(-1)**n)/(n**2)
    serie_truncada += coeficiente_a * np.cos(n*x)

# Graficar la función original y la serie truncada
plt.plot(x, f(x), label='Función original')
plt.plot(x, serie_truncada, label=f'Serie truncada ({num_terminos} términos)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.title('Convergencia de la serie de Fourier')
plt.grid(True)
plt.show()
```

```python
##10
import numpy as np
import matplotlib.pyplot as plt

# Definir la función original
def f(x):
    return x**2

# Definir el número máximo de armónicos
n_max = 10

# Definir el intervalo de valores de x
x = np.linspace(-np.pi, np.pi, 1000)

# Calcular los coeficientes a_n
a_n = np.zeros(n_max)
for n in range(1, n_max + 1):
    integrand = lambda x: f(x) * np.cos(n * x)
    a_n[n-1] = 1/np.pi * np.trapz(integrand(x), x)

# Calcular la serie truncada de Fourier
f_n = np.zeros_like(x)
for n in range(1, n_max + 1):
    f_n += a_n[n-1] * np.cos(n * x)

# Graficar la función original y la serie truncada
plt.plot(x, f(x), label='Función original')
plt.plot(x, f_n, label=f'Serie truncada ({n_max} armónicos)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.title('Convergencia de la serie de Fourier')
plt.grid(True)
plt.show()
```
